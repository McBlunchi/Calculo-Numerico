```python
def funcion(x):
    return np.sqrt(1+x)

def P_2funcion(x):
    return 1+x/2 - x**2 / 8

x = np.linspace(0,10,1000)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.plot(x, funcion(x), label='f(x)')
ax1.plot(x, P_2funcion(x), label='P_2(x)')
ax1.legend()

ax2.plot(x, (funcion(x) - P_2funcion(x)) / funcion(x), label='Error relativo')
ax2.legend()

plt.show()
print(funcion(2)-P_2funcion(2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def polinomio(a, b, c, x):
    return a * x**2 + b * x + c

a, b, c = 1, -40, 0.25

x = np.linspace(-10, 50, 100000)
f_x = polinomio(a, b, c, x)

target = 0
indices_closest = np.where(np.isclose(f_x, target, atol=0.04))[0]

x_closest = x[indices_closest]
f_x_closest = f_x[indices_closest]

plt.plot(x, f_x)
plt.axhline(0, color='red', linestyle='--', linewidth=0.8)
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Polinomio de segundo grado")

plt.scatter(x_closest, f_x_closest, color="green", label="Aprox. races")
plt.legend()
plt.tight_layout()
plt.show()

for i in range(len(x_closest)):
    print(f"Raz aproximada en x = {x_closest[i]} con f(x) = {f_x_closest[i]}")
```

```python
def raices_cuadraticas(a, b, c):
  discriminante = (b**2) - 4*(a*c)
  x1 = (-b + np.sqrt(discriminante)) / (2*a)
  x2 = (-b - np.sqrt(discriminante)) / (2*a)
  return x1, x2

x1, x2 = raices_cuadraticas(a, b, c)
print("Las races son:", x1, x2)
```

```python
def numeros_maquina(numero, m):
    if abs(numero) < 1:
        n_redondeado = round(abs(numero), m - len(str(int(abs(numero)))) + 1 )
    else:
        n_redondeado = round(abs(numero), m - len(str(int(abs(numero)))))

    return n_redondeado if numero >= 0 else -n_redondeado

print(numeros_maquina(0.1232512312, 4))
print(numeros_maquina(-123444432, 4))
```

```python
def raices(a, b, c, m):
    b_rounded = numeros_maquina(b, m)
    discriminant_rounded = numeros_maquina(np.sqrt(b**2 - 4*a*c), m)

    x1 = numeros_maquina((-b_rounded + discriminant_rounded) / (2*a), m)
    x2 = numeros_maquina((-b_rounded - discriminant_rounded) / (2*a), m)

    return x1, x2

raices(1,-40,0.25,4)
```
