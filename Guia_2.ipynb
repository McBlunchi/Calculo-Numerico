# Probando el Forzado

```python
def euler0(f, t0, tn, x_0, h, v_0, a, b, c, w):
    n = int((tn - t0) / h)
    t = np.linspace(t0, tn, n+1)
    x, v = np.zeros(n+1), np.zeros(n+1)

    x[0] = x_0
    v[0] = v_0

    for i in range(n):
        ff = f(t[i], x[i], v[i], a, b, c, w)
        f1_val, f2_val = ff

        x[i + 1] = x[i] + h * f1_val
        v[i + 1] = v[i] + h * f2_val

    return t, x, v

def forzado(t,x,v,a,b,c,w):
    return [v, -a*v-b*x+c*np.cos(w*t)]

h = 0.001
t0 = 0
tn = 1000
x_0 = 2
v_0 = 0
a = 0.2
b = 1
c = 100
w = 1

t1, x1, v1 =euler0(forzado, t0, tn, x_0, h, v_0, a,b,c,w)
```

```python
def SIR(t, y, beta, gamma, N):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]
```

```python
def evolSIR(SIR, t0, tf, y_0, h, beta, gamma, N):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n+1)
    s, i, r = np.zeros(n+1), np.zeros(n+1), np.zeros(n+1)
    s_0, i_0, r_0 = y_0

    s[0] = s_0
    i[0] = i_0
    r[0] = r_0


    for j in range(n):
        y = [s[j], i[j], r[j]]

        ff = SIR(t[j], y, beta, gamma, N)
        fs_val, fi_val, fr_val = ff

        s[j + 1] = s[j] + h * fs_val
        i[j + 1] = i[j] + h * fi_val
        r[j + 1] = r[j] + h * fr_val

    return t, s, i, r

h = 1e-3
beta = 0.2
gamma = 0.1
N = 10000
t0 = 0
tf = 100
y_0 = np.array([9999, 1, 0])
t, s, i, r = evolSIR(SIR, t0, tf, y_0, h, beta, gamma, N) # Now we are unpacking all 4 return values from evolSIR
```

```python
plt.plot(t, s, label='Susceptible')
plt.plot(t, i, label='Infected')
plt.plot(t, r, label='Recovered')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```
